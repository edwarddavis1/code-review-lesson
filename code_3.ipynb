```python
import numpy as np
from scipy.signal import argrelextrema

# Define the complicated function
def f(x):
    return np.sin(x) + np.sin(0.5 * x) + np.sin(0.2 * x)

# Generate x values
x = np.linspace(0, 20, 1000)

# Calculate y values
y = f(x)

# Find indices of relative minima
indices = argrelextrema(y, np.less)

# Print the x values corresponding to the troughs
troughs = x[indices]
print(troughs)
```

```python
import matplotlib.pyplot as plt

# Plot the function
plt.figure()
plt.plot(x, y)

# Plot the troughs
plt.scatter(troughs, f(troughs), c='r')
```
